```python
import pandas as pd
# print(pd.__version__)
import numpy as np

# ke numpy values easier to read.
np.set_printoptions(precision=3, suppress=True)
import tensorflow as tf
from tensorflow.keras import layers
```

```python
S_290922_20 = pd.read_csv(
    "/Users/JeroenArians/MakeAIWork/simpylc/default_samples_files/csv/20_S_290922.csv",
    names=["distance 01", "distance 02","distance 03", "angle"])

S_290922_20.head()
```

```python
distanceLearning = S_290922_20.copy()
learningWithoutAngle = distanceLearning.pop('angle')
learningWithoutAngle.head()
```

```python
distanceLearning = np.array(distanceLearning)
distanceLearning
```

```python
LearningModel = tf.keras.Sequential([
  layers.Dense(64, activation = 'elu'),
  layers.Dense(64, activation = 'elu'),
  layers.Dense(64, activation = 'relu'),
  layers.Dense(64, activation = 'relu'),  
  layers.Dense(34)
])

LearningModel.compile(loss = tf.keras.losses.MeanSquaredError(),
                      optimizer = tf.keras.optimizers.Adam(learning_rate=0.00001))
```

```python
LearningModel.fit(distanceLearning, learningWithoutAngle, epochs=500)
```
